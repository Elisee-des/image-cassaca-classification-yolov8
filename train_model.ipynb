```python
# Importer le module YOLO
from ultralytics import YOLO
import os
from pathlib import Path

DATA_DIR = '/home/safiatou/yolov8_env/image-cassaca-classification-yolov8/dataset'

def count_files_in_subdirectories(directory):
    file_counts = {}
    for root, dirs, files in os.walk(directory):
        for dir_name in dirs:
            dir_path = os.path.join(root, dir_name)
            file_count = len([f for f in os.listdir(dir_path) if os.path.isfile(os.path.join(dir_path, f))])
            subdir_key = os.path.relpath(dir_path, directory)
            file_counts[subdir_key] = file_count
    return file_counts

file_counts = count_files_in_subdirectories(DATA_DIR)

print("Nombre de fichiers dans chaque dossier :")
for dir_name, count in file_counts.items():
    print(f"{dir_name}: {count} fichiers")
```

```python
# Charger le modÃ¨le YOLOv8 pour la classification
model = YOLO('yolov8n-cls.pt')

# EntraÃ®ner le modÃ¨le
model.train(data=DATA_DIR, epochs=20, imgsz=224)

# Ã‰valuer le modÃ¨le sur l'ensemble de validation
metrics = model.val()

# Afficher les mÃ©triques d'Ã©valuation
print(metrics)
```
